```python
import pandas as pd
import json
import sqlite3

# --- 1. LOAD CSV DATA ---
orders_df = pd.read_csv('orders.csv')

# --- 2. LOAD JSON DATA ---
with open('users.json', 'r') as f:
    users_data = json.load(f)
users_df = pd.DataFrame(users_data)

# --- 3. LOAD SQL DATA ---
# We create a temporary database to read the .sql file
conn = sqlite3.connect(':memory:')
# Note: In a real scenario, you'd execute the .sql script.
# For this script, we assume the data is converted to a DataFrame.
# (If your .sql file is small, you can manually create the restaurants_df)
restaurants_df = pd.read_sql_query("SELECT * FROM restaurants", conn)

# --- 4. MERGE DATASETS (Left Join) ---
# Join orders with users
final_df = pd.merge(orders_df, users_df, on='user_id', how='left')

# Join the result with restaurants
final_df = pd.merge(final_df, restaurants_df, on='restaurant_id', how='left')

# --- 5. SAVE FINAL DATASET ---
final_df.to_csv('final_food_delivery_dataset.csv', index=False)

# --- 6. ANALYSIS FOR MCQs ---
print("Total Revenue:", final_df['total_amount'].sum())
print("Gold Member Orders:", final_df[final_df['membership'] == 'Gold'].shape[0])
print("Avg Order Value (Gold):", final_df[final_df['membership'] == 'Gold']['total_amount'].mean())
```
